The datasets used in this project include **Montgomery** and **Shenzhen**, which have been combined and split into train, validation, and test subsets. This ensures that the model generalizes well across different sources of X-ray images.

# üìÇ Splitting Dataset into Train, Validation, and Test

We split the dataset images into **train, validation, and test sets** using a 70-15-15% split:

```python
image_path = Path("dataset/img")
mask_path = Path("dataset/mask")

image_files = os.listdir(image_path)

train_files, val_files = train_test_split(image_files, test_size=0.3, random_state=42)
val_files, test_files = train_test_split(val_files, test_size=0.5, random_state=42)

print("train files:", len(train_files))
print("validation files:", len(val_files))
print("test files:", len(test_files))
```

# ü´Å Lung Dataset & Data Augmentation

This cell defines the **custom Dataset class** for lung segmentation and the **data augmentation**:

- üóÇÔ∏è `LungDataset`: Loads images and their corresponding masks, normalizes them, and applies optional transformations.  
- üé® `train_transform`: Performs resizing, flips, rotations, shift/scale/rotate, and brightness/contrast adjustments for robust training.  
- üîπ `val_transform`: Only resizes and converts to tensor for validation, keeping data consistent.

This setup ensures the model sees **diverse examples** during training while validation data remains unaltered.


```python
class LungDataset(Dataset):
    def __init__(self, image_dir, mask_dir, file_list, transform=None):
        self.image_dir = image_dir
        self.mask_dir = mask_dir
        self.file_list = file_list
        self.transform = transform

    def __len__(self):
        return len(self.file_list)

    def __getitem__(self, index):
        file_name = self.file_list[index]

        
        image_path = os.path.join(self.image_dir, file_name)
        image = np.array(Image.open(image_path).convert('L'), dtype=np.float32) / 255.0
        image = np.expand_dims(image, axis=-1)  # (H,W,1)

        
        mask_path = os.path.join(self.mask_dir, file_name)
        mask = np.array(Image.open(mask_path).convert('L'), dtype=np.float32) / 255.0
        mask = np.expand_dims(mask, axis=-1)  # (H,W,1)

        
        if self.transform:
            augmented = self.transform(image=image, mask=mask)
            image = augmented['image']      # (1,H,W)
            mask  = augmented['mask']       # (1,H,W)
        else:
            image = torch.from_numpy(image).permute(2,0,1)
            mask  = torch.from_numpy(mask).permute(2,0,1)

        return image.float(), mask.float()

# =========================
# Transforms
# =========================
train_transform = A.Compose([
    A.Resize(256, 256),
    A.HorizontalFlip(p=0.5),
    A.VerticalFlip(p=0.5),
    A.RandomRotate90(p=0.5),
    A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=15, p=0.5),
    A.RandomBrightnessContrast(p=0.5),
    ToTensorV2()
])

val_transform = A.Compose([
    A.Resize(256, 256),
    ToTensorV2()
])
```

# üöÄ Creating Datasets and DataLoaders

We create **train, validation, and test datasets** using the `LungDataset` class and wrap them in **DataLoaders**:  

- üèãÔ∏è‚Äç‚ôÇÔ∏è `trainLoader`: Batch size 2, shuffling enabled for training  
- üîç `valLoader`: Batch size 2 for validation  
- üß™ `testLoader`: Batch size 4 for testing  

This prepares the data for **efficient model training and evaluation**.


```python
trainDataset = LungDataset(image_dir=image_path, mask_dir=mask_path, file_list=train_files, transform=train_transform)
validationDataset = LungDataset(image_dir=image_path, mask_dir=mask_path, file_list=val_files, transform=val_transform)
testDataset = LungDataset(image_dir=image_path, mask_dir=mask_path, file_list=test_files, transform=val_transform)
trainLoader = DataLoader(trainDataset, batch_size=2, shuffle=True)
valLoader = DataLoader(validationDataset, batch_size=2)
testLoader = DataLoader(testDataset, batch_size=4)
```

# ü§ñ Defining the UNet++ Model

We initialize a **UNet++** model for lung segmentation:  

- üß† **Encoder**: EfficientNet-B0 pre-trained on ImageNet  
- ü´Å **Input channels**: 1 (grayscale X-ray images)  
- üéØ **Output classes**: 1 (binary mask)  

We then use `summary` to **inspect the model architecture and parameter count**.  


```python
model = smp.UnetPlusPlus(
    encoder_name="efficientnet-b0",
    encoder_weights="imagenet",
    in_channels=1,   
    classes=1        
).to("cuda" if torch.cuda.is_available() else "cpu")

summary(model, input_size=(1, 1, 256, 256))
```

# ‚ö° Loss Function & Optimizer

- üßÆ **Losses**:  
  - `BCEWithLogitsLoss` for pixel-wise binary classification  
  - `DiceLoss` for overlap measurement  
  - ‚úÖ `combined_loss` = BCE + Dice  

- ‚öôÔ∏è **Optimizer**: Adam with learning rate 1e-3  
- üìâ **Scheduler**: ReduceLROnPlateau to reduce LR if validation loss plateaus


loss_dice = smp.losses.DiceLoss(mode='binary', from_logits=True)
loss_bce = nn.BCEWithLogitsLoss()

def combined_loss(prediction, target):
    return loss_bce(prediction, target) + loss_dice(prediction, target)

optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
scheduler = ReduceLROnPlateau(optimizer, mode='min', patience=5, factor=0.1, verbose=True)

# üéØ Dice Score

Computes the **Dice Score** for segmentation:  
- üîπ Converts predictions to binary masks  
- üîπ Measures overlap between prediction and ground truth (0‚Äì1, 1 = perfect)


```python
def dice_score(preds, targets, threshold=0.5):
    preds = torch.sigmoid(preds)            
    preds = (preds > threshold).float()     
    targets = targets.float()
    
    intersection = (preds * targets).sum(dim=(1,2,3))
    union = preds.sum(dim=(1,2,3)) + targets.sum(dim=(1,2,3))
    
    dice = (2. * intersection + 1e-6) / (union + 1e-6)
    return dice.mean().item()
```

# üöÄ Training Function

Defines `train_model` to train and validate the UNet++ model:

- üèãÔ∏è‚Äç‚ôÇÔ∏è **Training Loop**: updates weights using `combined_loss`  
- üîç **Validation Loop**: evaluates loss and Dice Score on validation set  
- üíæ **Best Model Saving**: saves model with highest validation Dice  
- üìä **History Tracking**: records train/val loss and Dice for plotting  
- üìâ **Scheduler**: updates learning rate based on validation loss


```python
def train_model(model, train_loader, val_loader, optimizer, scheduler, num_epochs, device):
    history = {"train_loss": [], "val_loss": [], "train_dice": [], "val_dice": []}
    best_val_dice = 0.0

    for epoch in range(num_epochs):
        # =====================
        # Training
        # =====================
        model.train()
        running_loss, running_dice = 0.0, 0.0

        train_bar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs} - Training", leave=False)
        for images, masks in train_bar:
            images = images.to(device)
            masks = masks.to(device)
            masks = masks.permute(0, 3, 1, 2)  # (B,H,W,1) -> (B,1,H,W)

            optimizer.zero_grad()
            outputs = model(images)

            loss = combined_loss(outputs, masks)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()
            running_dice += dice_score(outputs, masks)

            train_bar.set_postfix({"Loss": running_loss/(train_bar.n+1), 
                                   "Dice": running_dice/(train_bar.n+1)})

        train_loss = running_loss / len(train_loader)
        train_dice = running_dice / len(train_loader)

        # =====================
        # Validation
        # =====================
        model.eval()
        val_running_loss, val_running_dice = 0.0, 0.0

        val_bar = tqdm(val_loader, desc=f"Epoch {epoch+1}/{num_epochs} - Validation", leave=False)
        with torch.no_grad():
            for images, masks in val_bar:
                images = images.to(device)
                masks = masks.to(device)
                masks = masks.permute(0, 3, 1, 2)

                outputs = model(images)
                loss = combined_loss(outputs, masks)

                val_running_loss += loss.item()
                val_running_dice += dice_score(outputs, masks)

                val_bar.set_postfix({"Loss": val_running_loss/(val_bar.n+1), 
                                     "Dice": val_running_dice/(val_bar.n+1)})

        val_loss = val_running_loss / len(val_loader)
        val_dice = val_running_dice / len(val_loader)

        # ===== Save best model =====
        if val_dice > best_val_dice:
            best_val_dice = val_dice
            torch.save(model.state_dict(), "best_model.pth")
            print(f"*** Saved Best Model at Epoch {epoch+1} - Val Dice: {val_dice:.4f} ***")


        history["train_loss"].append(train_loss)
        history["val_loss"].append(val_loss)
        history["train_dice"].append(train_dice)
        history["val_dice"].append(val_dice)

        
        print(f"Epoch [{epoch+1}/{num_epochs}] "
              f"- Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f} "
              f"- Train Dice: {train_dice:.4f}, Val Dice: {val_dice:.4f}")

        
        scheduler.step(val_loss)

    return history
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)
NUM_EPOCHS = 30

history = train_model(
    model=model,
    train_loader=trainLoader,
    val_loader=valLoader,
    optimizer=optimizer,
    scheduler=scheduler,
    num_epochs=NUM_EPOCHS,
    device=device
)
```

# üß™ Testing the Model

This function evaluates the **best saved model** on the test dataset:

- üîπ Loads `best_model.pth`  
- üîπ Computes **loss** and **Dice Score** for each test image  
- üîπ Displays a few samples with:  
  - Original image üñºÔ∏è  
  - Ground truth mask üü©  
  - Predicted mask üîπ  
- üìä Prints **average test loss and Dice Score** for overall performance.


```python
def test_model(model, test_dataset, device, num_samples=5):
    model.load_state_dict(torch.load("best_model.pth", map_location=device, weights_only=True))
    model.to(device)
    model.eval()

    test_loader = DataLoader(test_dataset, batch_size=1, shuffle=False)
    total_dice = 0.0
    total_loss = 0.0

    samples_shown = 0

    with torch.no_grad():
        for images, masks in test_loader:
            images = images.to(device)
            masks = masks.to(device)
            masks = masks.permute(0, 3, 1, 2)

            outputs = model(images)
            loss = combined_loss(outputs, masks)
            total_loss += loss.item()
            batch_dice = dice_score(outputs, masks)
            total_dice += batch_dice

            if samples_shown < num_samples:
                pred_mask = torch.sigmoid(outputs).cpu().numpy()[0,0]
                true_mask = masks.cpu().numpy()[0,0]
                image = images.cpu().numpy()[0,0]

                sample_dice = dice_score(outputs, masks)  # ŸáŸÖ€åŸÜÿ¨ÿß item ÿ≠ÿ∞ŸÅ ÿ¥ÿØ
                print(f"Sample {samples_shown+1} - Dice Score: {sample_dice:.4f}")

                plt.figure(figsize=(12,4))
                plt.subplot(1,3,1)
                plt.imshow(image, cmap='gray')
                plt.title("Image")
                plt.axis('off')

                plt.subplot(1,3,2)
                plt.imshow(true_mask, cmap='gray')
                plt.title("Ground Truth Mask")
                plt.axis('off')

                plt.subplot(1,3,3)
                plt.imshow(pred_mask > 0.5, cmap='gray')
                plt.title("Predicted Mask")
                plt.axis('off')

                plt.show()
                samples_shown += 1

    avg_loss = total_loss / len(test_loader)
    avg_dice = total_dice / len(test_loader)
    print(f"Test Loss: {avg_loss:.4f}, Test Dice: {avg_dice:.4f}")
```

```python
test_model(model, testDataset, device, num_samples=3)
```
